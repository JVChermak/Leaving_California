```python
# Try to import affordability table
# Names array
names = ["Region_division_area", "Total", "Total_Number_Cant_Afford_Med_home", "Total_Percent_Cant_Afford_Med_Home", "Total_Curr_Owners", "Total_Number_Curr_Own_Cant_Afford_Med", "Total_Percent_Curr_Own_Cant_Afford_Med", "Total_Current_Renters", "Total_Num_Curr_Rent_Cant_Afford_Med", "Total_Percent_Curr_Rent_Cant_Afford_Med"]
affordability_2009 = pd.read_excel("Resources/tab1-3_web.xls", header=7, names=names, index_col=None, skiprows=2, nrows=14)
affordability_2009
```

```python
# Remove periods from first column
regions = affordability_2009['Region_division_area']
regions
```

```python
import re
import numpy as np

form_one = r'\w+\s+\w+\.*'
form_two = r'\.{2}\w+\s+\w+\.*'
form_three = r'\.{2}\w+\.*'
form_four = r'\.{2}\w+\s+\w+\s+\w+\.*'
form_five = r'\w+\.*'
def parse_regions(region):
    if type(region) != str:
        return np.nan
    if re.match(r'\w+\s+\w+\.*', region, flags=re.IGNORECASE):
        region = re.sub(r'\.+', '', region)
        return region
    elif re.match(r'\.{2}\w+\s+\w+\.*', region, flags=re.IGNORECASE):
        region = re.sub(r'\.+', '', region)
        return region
    elif re.match(r'\.{2}\w+\.*', region, flags=re.IGNORECASE):
        region = re.sub(r'\.+', '', region)
        return region
    elif re.match(r'\.{2}\w+\s+\w+\s+\w+\.*', region, flags=re.IGNORECASE):
        region = re.sub(r'\.+', '', region)
        return region
    elif re.match(r'\w+\.*', region, flags=re.IGNORECASE):
        region = re.sub(r'\.+', '', region)
        return region
    else:
        return np.nan
affordability_2009['Region'] = regions.str.extract(f'({form_one}|{form_two}|{form_three}|{form_four}|{form_five})', flags=re.IGNORECASE)[0].apply(parse_regions)
```

```python
affordability_2009.drop(columns=['Region_division_area'], inplace=True)
cols = affordability_2009.columns.tolist()
cols = cols[-1:] + cols[:-1]
affordability_2009 = affordability_2009[cols]
```
