```python
# Import our dependencies
import pandas as pd
import numpy as np
```

```python
# Read in our data
df = pd.read_csv("../Resources/clean_data/Merged/region_migration.csv")
df.head()
```

```python
# Make sure dtypes are set up for analysis
df.dtypes
```

```python
# Convert Year to datetime
df['Year'] =  pd.to_datetime(df['Year'], format='%Y')
```

```python
# Groupby Region to get total Californians moved to each region
df = df.groupby(['NAME']).sum()
df
```

```python
df = df.sort_values(by='value', axis=0, ascending=False, inplace=False, kind='quicksort', na_position='last')
df.head()
```

By grouping the dataframe by state, sum the values, and sorting the results in descending order: We can see the top 5 states Californian's are moving to are Texas, Arizona, Washington, New York, and Nevada.

```python
# Saving cleaned data
file_path = "../Resources/analysis/state_migration.csv"
df.to_csv(file_path, index=False)
```
