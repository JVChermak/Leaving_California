```python
# Import our dependencies
import pandas as pd
import numpy as np
```

```python
# Read in our data
df = pd.read_csv("../Resources/clean_data/Migration/ca_migration_2010.csv")
df.head()
```

```python
df = pd.melt(df, id_vars=['Year'])
df
```

```python
# Read in our data
pop = pd.read_csv("../Resources/clean_data/Population/pop_2010.csv")
pop.head()
```

```python
pop.shape
```

```python
# Only view needed column information
pop = pop.filter(items=['Region','NAME'])
pop
```

```python
# Change column names
df.rename(columns={'variable': 'NAME'}, inplace=True)
df.head()
```

```python
mig = pd.merge(df, pop, on='NAME', how='right').fillna(method='ffill')
```

```python
mig.shape
```

```python
mig.head(55)
```

```python
# Reorder columns
mig = mig[['Year', 'Region', 'NAME', 'value']]
mig.head()
```

```python
# Saving cleaned data
file_path = "../Resources/clean_data/Merged/migration_2010.csv"
mig.to_csv(file_path, index=False)
```
