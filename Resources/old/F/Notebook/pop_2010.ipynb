```python
# Import our dependencies
import pandas as pd
```

```python
# Read in our data
df = pd.read_csv("../Resources/clean_data/population_data.csv")
df.head()
```

```python
df.columns
```

```python
# Only view needed column information
pop = df.filter(items=['Region', 'NAME', 'Population 2010', 'Net Pop Chg 2010', 'Net Migration 2010', '%Net Migration 2010'])
pop
```

```python
# Adding 'Year' column
pop.insert(0, 'Year', 2010)
pop.head()
```

```python
# Change column names
pop.rename(columns={'Population 2010': 'Population', 'Net Pop Chg 2010': 'Net Population Change', 'Net Migration 2010': 'Net Migration', '%Net Migration 2010': 'Net Migration Change'}, inplace=True)
pop.head()
```

```python
pop.dtypes
```

```python
# Saving cleaned data
file_path = "../Resources/clean_data/pop_2010.csv"
pop.to_csv(file_path, index=False)
```
