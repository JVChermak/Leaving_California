```python
# Import our dependencies
import pandas as pd
import numpy as np
```

```python
# Read in our data
census = pd.read_csv("../Resources/clean_data/census_2017.csv")
```

```python
# Read in our data
pop = pd.read_csv("../Resources/clean_data/pop_2017.csv")
```

```python
# Change column names
census.rename(columns={'State': 'NAME'}, inplace=True)
```

```python
df = pd.merge(census, pop, on='NAME', how='left').fillna(method='ffill')
```

```python
# Change column names
df.rename(columns={'Year_x': 'Year'}, inplace=True)
```

```python
# Delete Year_y column 
del df['Year_y']
```

```python
# Saving cleaned data
file_path = "../Resources/clean_data/2017.csv"
df.to_csv(file_path, index=False)
```
