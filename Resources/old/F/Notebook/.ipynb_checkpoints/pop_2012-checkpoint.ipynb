```python
# Import our dependencies
import pandas as pd
```

```python
# Read in our data
df = pd.read_csv("../Resources/clean_data/population_data.csv")
df.head()
```

```python
df.columns
```

```python
# Only view needed column information
pop = df.filter(items=['Region', 'NAME', 'Population 2012', 'Net Pop Chg 2012', 'Net Migration 2012', '%Net Migration 2012'])
pop.head()
```

```python
# Adding 'Year' column
pop.insert(0, 'Year', 2012)
pop.head()
```

```python
# Change column names
pop.rename(columns={'Population 2012': 'Population', 'Net Pop Chg 2012': 'Net Population Change', 'Net Migration 2012': 'Net Migration', '%Net Migration 2012': 'Net Migration Change'}, inplace=True)
pop.head()
```

```python
pop.dtypes
```

```python
# Saving cleaned data
file_path = "../Resources/clean_data/pop_2012.csv"
pop.to_csv(file_path, index=False)
```
