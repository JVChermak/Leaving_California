#### Data  Analysis on Master

```python
# Statistics 
# df['column'].describe()

# Group data by region and summarize "Different state of residence (1yr ago) Total Est"
x= Master_CleanData_Migration_df2.groupby(["Region", "Year"])[["Different state of residence (1yr ago) Total Est"]].describe()
#x.sort_values(by='Year', 'Region', ascending=False)

x.sort_values(['Year', 'Region'], ascending=[True, False])
```

```python
# Total Migration "Out"-flow per State per Year
Migration_per_State_Count = Master_CleanData_Migration_df2.set_index(["Year", "Current residence"])["Different state of residence (1yr ago) Total Est"]
Migration_per_State_Count
```

```python
# Calculate the mean migration per region.
per_region = Master_CleanData_Migration_df2.groupby([ "Year", 'Region'])[["Different state of residence (1yr ago) Total Est"]].mean()
per_region
```

```python
# Calculate the sum migration per region.
per_region = Master_CleanData_Migration_df2.groupby([ "Year", 'Region'])[["Different state of residence (1yr ago) Total Est"]].sum()
per_region
```

```python
# How many CA migrations per Region
per_region_from_CA = Master_CleanData_Migration_df2.set_index(['Year', 'Region'])["California"]
per_region_from_CA= Master_CleanData_Migration_df2.groupby(['Year', 'Region'])[["California"]].sum()

# Convert series to DF
df1 = pd.DataFrame(per_region_from_CA)
df1
```

```python
# Calculate  Top 5 states/by year with highest migration flows 
highest_mig2010 = Master_CleanData_Migration_df2[(Master_CleanData_Migration_df2["Different state of residence (1yr ago) Total Est"] >= 400000)]
highest_mig2010 
```

```python
# CA migration flow to all other states
CA_df = Master_CleanData_Migration_df2.loc[Master_CleanData_Migration_df2['Current residence'] == "California"]

# #Drop Unneeded columns for this analysis
CA_df2 = CA_df.drop(['Region', 'Population (1yr ago) Est','Same house (1yr ago) Est',
                                          'Same state of residence (1yr ago) Est' ,'Different state of residence (1yr ago) Total Est'], axis=1)

# reset Index
#CA_df .reset_index(drop=True, inplace=True)
CA_df2.sort_values(by='California', ascending=False)
CA_df2
```

```python
# # CA migration flow to all other states
Cali_Analysis_df = CA_df2.loc[CA_df2['Current residence'] == "California"]

# reset Index
Cali_Analysis_df.reset_index(drop=True, inplace=True)
Cali_Analysis_df.sort_values(by='California', ascending=False)
Cali_Sort= Cali_Analysis_df.transpose()
Cali_Sort
```

```python
Cali_Sort.columns = ['2018', '2017', '2016', '2015' , '2014', '2013', '2012', '2011', '2010']
Cali_Sort
```

```python
# key = column when changed into dict, then insert into DB
sorted_cali = Cali_Sort
```
